### Buat DataFrame yang mendeskripsikan Tiap Tiap kolom dari data Startup Funding

```python
ListFitur = list(df.columns)
ListType = []
ListNull = []
ListPct = []
ListUniq = []
ListUniqs = []
for item in ListFitur:
    ListType.append(df[item].dtype)
    ListNull.append(df[item].isna().sum())
    ListPct.append(round(df[item].isna().sum() / len(df)) * 100)
    ListUniq.append(df[item].nunique())
    ListUniqs.append(list(df[item].value_counts().head(2).sort_index(ascending = True).index))

data = {
    'dataFeatures' : ListFitur,
    'dataType' : ListType,
    'null' : ListNull,
    'nullPct': ListPct,
    'unique': ListUniq,
    'uniqueSample' : ListUniqs
}
pd.DataFrame(data)
```

####  lelaki yang merokok  VS  Perempuan yang merokok

```python
x = df[df['smoker'] == 'Yes']
for item in list(x['sex'].unique()):
        y = len(x[x['sex'] == item])
        print(item + ' smoker : ' + str(y))
```

####  mean tip dari lelaki yang merokok vs mean tip dari lelaki yang tidak merokok

```python
x = df[df['sex'] == 'Male']
for item in list(x['smoker'].unique()):
        y = x[x['smoker'] == item]['tip'].mean()
        print(str(y))
```

####  mean tip dari perempuan vs mean tip dari lelaki 

```python
for item in list(df['sex'].unique()):
        y = df[df['sex'] == item]['tip'].mean()
        print('Avg tip from ' + item + ' = ' + str(y))
```

#### Buatlah dataFrame seperti dibawah?

```python
hari = list(df['day'].unique())
pengunjung = []
totalTip = []
percTip = []
meanTip = []

for item in hari:
    pengunjung.append(df[df['day'] == item]['size'].sum())
    totalTip.append(df[df['day'] == item]['total_bill'].count())
    percTip.append(round((df[df['day'] == item]['tip'].count()) / (df[df['day'] == item]['size'].sum()) * 100))
    meanTip.append(round(df[df['day'] == item]['tip'].mean()))
    
data = {
    'hari' : hari,
    'totalPengunjung' : pengunjung,
    'totalPemberiTip' : totalTip,
    'percPemberiTip' :  percTip,
    'meanTip' : meanTip
}

pd.DataFrame(data)
```

#### Buatlah DataFrame Seperti dibawah ini

```python
waktu= list(df['time'].unique())
pengunjung = []
totalTip = []
percTip = []
meanTip = []

for item in waktu:
    pengunjung.append(df[df['time'] == item]['size'].sum())
    totalTip.append(df[df['time'] == item]['total_bill'].count())
    percTip.append(round((df[df['time'] == item]['tip'].count()) / (df[df['time'] == item]['size'].sum()) * 100))
    meanTip.append(round(df[df['time'] == item]['tip'].mean()))
    
data = {
    'waktu' : waktu,
    'totalPengunjung' : pengunjung,
    'totalPemberiTip' : totalTip,
    'percPemberiTip' :  percTip,
    'meanTip' : meanTip
}

pd.DataFrame(data)
```

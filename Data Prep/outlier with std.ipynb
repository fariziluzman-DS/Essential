```python
def detect_outlier(col) :
    std = data[col].std()
    lowerlimit = data[col].mean() - (3*std)
    upperlimit = data[col].mean() + (3*std)
    return len(data[(data[col]<lowerlimit) | (data[col]>upperlimit)])
```

```python
detect_outlier('age')
```

```python
def detect_outlier_return_index(col) :
    std = data[col].std()
    lowerlimit = data[col].mean() - (3*std)
    upperlimit = data[col].mean() + (3*std)
    return data[(data[col]<lowerlimit) | (data[col]>upperlimit)].index
```

```python
detect_outlier_return_index('fare')
```

```python
idoutlier=list(detect_outlier_return_index('fare'))
```

```python
data['fare'].mean()
```

```python
def applyfareoutlier(x) :
    if x['index'] in idoutlier :
        return data['fare'].mean()
    else :
        return x['fare']
```

```python
data['index']=data.index
```

```python
data[['index','fare']].apply(applyfareoutlier,axis=1)
```
